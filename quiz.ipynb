```python
import matplotlib.pyplot as plt

import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["CSE", "Mech", "Civil", "ECE"]
plt.pie(y, labels = mylabels)
plt.legend()
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels = ["Python", "chemisrty", "English", "Maths"]
myexplode = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```
